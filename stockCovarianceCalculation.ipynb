```python

import pandas as pd
import yfinance as yf
import numpy as np
import datetime
```

```python
sp_51 = pd.read_csv('SP500_preparation.csv',index_col= [0]) # the dataframe got from the SP500_preparation.ipynb
names = sp_51.index

s = names.size
```

Covariance matrix

```python
# set starting and ending time

start = datetime.datetime(2020,10,1)
end = datetime.datetime(2021,10,1)

# make a dict: key is the name of stock, value is an array of monthly increasing rate
sp_50_dict = {}
```

```python
sp_50_cov = np.zeros((s,s)) # covariance mat
```

```python
# build covariance matrix
def calculate_increases(name, start, end, period):
    try:
        prices = yf.download(name, start, end)
    except:
        return None
    
    Endings = prices.Close
    for i in range(len(Endings)):
        if Endings[i] is np.nan: Endings[i] = Endings[i-1]    
    increases = []
    for i in range(len(Endings) - period):
        increase = (Endings[i+period] - Endings[i])/Endings[i]
        increases.append(increase)
    increases_ = np.array(increases)
    return increases

for name in names:
    sp_50_dict[name] = calculate_increases(name, start, end, 20)

for i in range(s):
    for j in range(s):
        cov = np.cov(sp_50_dict[names[i]], sp_50_dict[names[j]])[0][1]
        sp_50_cov[i,j] = cov
```

```python
covariance_df = pd.DataFrame(sp_50_cov)


covariance_df.to_csv('49StocksCovariance.csv', index=False)

covariance_df
```
